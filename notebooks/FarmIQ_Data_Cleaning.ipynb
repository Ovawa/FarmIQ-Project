Removing any outliers in our numerical values

```python
# Function to remove outliers using IQR
def remove_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    cleaned = data[(data[column] >= lower) & (data[column] <= upper)]
    print(f"'{column}'  removed {len(data) - len(cleaned)} outliers")
    return cleaned

# Apply to key numeric columns
for col in ['Rainfall', 'Temperature', 'Ph', 'Production']:
    df = remove_outliers_iqr(df, col)

# Check final shape
print("Final shape after outlier removal:", df.shape)
```

Encoding crop type for Model to understand.

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Crop_encoded'] = le.fit_transform(df['Crop'])

# See the unique crop names and codes
crop_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
print(" Crop mapping:", crop_mapping)
#45 different kinds of crop found
```
